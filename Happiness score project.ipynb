```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x = pd.DataFrame(st.fit_transform(x), columns = x.columns)
x
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Checking for the best random_state value
```

```python
maxacc=0
maxrs=0
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    RFR=RandomForestRegressor()
    RFR.fit(x_train,y_train)
    pred=RFR.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i

print('max accuracy is :',maxacc,'random state :',maxrs)  
```

```python
# splitting our dataset based on the best random_state value
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=maxrs)
```

```python
# Importing regression algorithms
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# making their instances
```

```python
LR=LinearRegression()
RD=Ridge()
LS=Lasso()
EN=ElasticNet()
svr=SVR()
DTR = DecisionTreeRegressor()
GB=GradientBoostingRegressor()
RFR=RandomForestRegressor()
```

```python
# Creating a loop with the instances
```

```python
loop=[LR,RD,LS,EN,svr,DTR,GB,RFR]
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# In this loop I am training my model for each algorithmn to check the r2_score,mean_abs_error,mean_squared_error,root mean squared error, cross_val_score and the difference between mean cross_val_score and r2_score
```

```python
for i in loop:
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print('This is for :',i)
    print('r2_score will be :',r2_score(y_test,pred))
    print('mean abs error will be :',mean_absolute_error(y_test,pred))
    print('mean squared error will be :',mean_squared_error(y_test,pred))
    print('root mean squared error will be :',np.sqrt(mean_squared_error(y_test,pred)))
    print('\n')
    score=cross_val_score(i,x,y,cv=5)
    print(score)
    print('mean CV :',score.mean())
    print("difference between mean CV and r2_score will be :",r2_score(y_test,pred)-score.mean())
    print('\n')
```

```python
# Here Linear Regression model has the best performance and less overfitting
```

```python
# Lets do the hyper parameter tuning by using grid search cv
```
